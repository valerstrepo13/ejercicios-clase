```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6.5, 0.1)
y = np.sin(x)
plt.plot(x, y, 'm--')
plt.xlabel("Ángulo (rad)")
plt.ylabel("Sin(x)")
plt.title("Función seno")
plt.xlim(min(x), max(x))
plt.ylim(min(y), max(y))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6.5, 0.1)
y = np.sin(x)
z = np.cos(x)
plt.plot(x, y, 'm--', label='Función seno')
plt.plot(x, z, 'g-', label='Función coseno')
plt.xlabel("Ángulo (rad)")
plt.legend()
plt.xlim(min(x), max(x))
plt.ylim(min(y), max(y))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y1 = 2 * x + 1
y2 = -x**2
plt.plot(x, y1, label='2x + 1')
plt.plot(x, y2, label='-x^2')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: vector de tamaño 720 con valores aleatorios
x = np.random.rand(720)

# Paso 2: Hacer reshape a (120, 6). Paso 3: transpuesta de la matriz y dos copias, una usando order F y la otra usando order C
y = x.reshape(120, 6)
y_transpuesta = y.T

matriz_F = np.copy(y_transpuesta, order='F')
matriz_C = np.copy(y_transpuesta, order='C')

# Paso 3: Crear un subplot con 6 paneles con diferentes proporciones "a mano"
fig, axs = plt.subplots(3, 2, figsize=(10, 8))

# Paso 4: En cada panel hacer un gráfico diferente
# Gráfico 1: Plot
axs[0, 0].plot(y[0])
axs[0, 0].set_title('Plot')
axs[0, 0].set_xlabel('Índice')
axs[0, 0].set_ylabel('Valor')

# Gráfico 2: Scatter
axs[0, 1].scatter(range(len(y[1])), y[1])
axs[0, 1].set_title('Scatter')
axs[0, 1].set_xlabel('Índice')
axs[0, 1].set_ylabel('Valor')

# Gráfico 3: Bar
axs[1, 0].bar(range(len(y[2])), y[2])
axs[1, 0].set_title('Bar')
axs[1, 0].set_xlabel('Índice')
axs[1, 0].set_ylabel('Valor')

# Gráfico 4: Histograma
axs[1, 1].hist(y[3], bins=3, color='red')
axs[1, 1].set_title('Histograma')
axs[1, 1].set_xlabel('Valor')
axs[1, 1].set_ylabel('Frecuencia')

# Gráfico 5: Pie
axs[2, 0].pie(y[4], labels=range(len(y[4])))
axs[2, 0].set_title('Pie')

# Gráfico 6: Boxplot
axs[2, 1].boxplot(y[5])
axs[2, 1].set_title('Boxplot')
axs[2, 1].set_xlabel('Índice')
axs[2, 1].set_ylabel('Valor')

for ax in axs.flat:
    ax.grid(True)

plt.tight_layout()
plt.show()

```
